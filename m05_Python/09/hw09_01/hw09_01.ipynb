**ЗАДАЧА 2.**
Узнать какая максимальная households в зоне минимального значения population

```python
# Выборка двух столбцов
df_populationHouseholds = df[['population', 'households']]
df_populationHouseholds
```

```python
# Статистика по ним
df_populationHouseholds.describe()
```

```python
# Минимальная и максимальная популяции
populMin = df_populationHouseholds['population'].min()
populMax = df_populationHouseholds['population'].max()
populMin, populMax
```

Дадим оценку термину "зона минимального значения population".
Видим, что min = 3, max = 35682.
Разберем случай, когда популяция < 1000, а потом в цикле прогоним несколько вариантов

```python
# Выборка, где популяция < 1000
min = 1000
df_min = df[df['population'] < min][['population', 'households']]
df_min
```

```python
# В таком случае максимальная households
householdsMax = df_min['households'].max()
householdsMax
```

Теперь циклом разберем много вариантов

```python
householdsMaxDict = {}
householdsMaxDict[min] = householdsMax
print(f"В зоне значений population до {min}, максимальная households = {householdsMaxDict[min]}")
popul = 2500
while (popul < populMax):
  df_popul = df[df['population'] < popul][['population', 'households']]
  householdsMaxDict[popul] = df_popul['households'].max()
  print(f"В зоне значений population до {popul}, максимальная households = {householdsMaxDict[popul]}")
  popul += 2500
```
