Новая библиотека (графики чертит)

```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins
```

Точечный график

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g")
```

Точечный график с изменением цвета

```python
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g")
```

Линейный график

```python
sns.relplot(x="bill_length_mm", y="bill_depth_mm", kind="line", data=penguins)
```

Гистограмма

```python
sns.histplot(data=penguins, x="body_mass_g")
```
